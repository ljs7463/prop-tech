```python
sql = f"""
SELECT danji.complexNo, danji.complexName, COUNT(price.dealPrice) AS cnt , max(totalHouseholdCount) AS totalHouseholdCount
FROM `aidepartners.aide.complex_real_price_prod` AS price
LEFT JOIN (
           SELECT complexNo, complexName, sidoCode, useApproveYear, totalHouseholdCount, realEstateTypeCode
           FROM `aidepartners.aide.complex_danji_information` 
           ) AS danji ON price.complexNo = danji.complexNo
WHERE danji.sidoCode = '11'
      AND danji.totalHouseholdCount >=200
      AND price.tradeType = 'A1' 
      AND price.deleteYn is null 
      AND danji.realEstateTypeCode = 'APT'
      AND price.contractDate >='2023-01-01'
GROUP BY danji.complexNo, danji.complexName
ORDER BY cnt DESC

"""
df = BH.read_table(sql)
```

```python
df_heatmap = df[['complexName','cnt','totalHouseholdCount']]
```

```python
df_heatmap['rotationRate'] = df_heatmap['cnt']/ df_heatmap['totalHouseholdCount']
df_heatmap['rotationRate'] = df_heatmap['rotationRate']*100
```

```python
df_1 = df_heatmap.sort_values(by= 'totalHouseholdCount',ascending = False).iloc[:25]
df_2 = df_heatmap.sort_values(by= 'rotationRate',ascending = False).iloc[:25]
```
