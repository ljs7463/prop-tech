```python
# qeury
sql = f"""
SELECT T2.*
FROM(
  SELECT T1.*,
        -- 단지별 유형별 대표 면적 (대표 면적 -> 유형별 타입별 세대수 가장많은 타입)
        ROW_NUMBER() OVER(PARTITION BY T1.complexNo, T1.areaCategory ORDER BY T1.householdCountByPyeong DESC) AS rn
  FROM(
      SELECT danji.complexNo, 
            danji.complexName, 
            danji.sigunguName, 
            danji.bdongName,
            danji.totalHouseholdCount,
            danji.useApproveYear,
            danji.HeatMethodTypeName,
            danji.constructionCompanyName,
            area.supplyArea,
            area.jeonyongArea,
            area.householdCountByPyeong,
            area.entranceType,
            area.roomCount,
            area.bathroomCount,
            area.pyeongNo,
            price.tradeType,
            price.dealPrice,
            price.leasePrice,
            price.rentPrice,
            price.contractDate,
            CASE 
            WHEN area.jeonyongArea >=15 AND area.jeonyongArea<=30 THEN '소형'
            WHEN area.jeonyongArea >=30 AND area.jeonyongArea<=45 THEN '중소형'
            END AS areaCategory

      FROM `aidepartners.aide.complex_danji_information` AS danji
      LEFT JOIN (
                  SELECT complexNo,
                        supplyArea,
                        jeonyongArea,
                        householdCountByPyeong,
                        entranceType,
                        roomCount,
                        bathroomCount,
                        pyeongNo
                  FROM `aidepartners.aide.complex_area_information_prod` 
                  ) AS area 
                    ON danji.complexNo = area.complexNo
      LEFT JOIN (
                  SELECT complexNo,
                        areaNo,
                        contractDate,
                        tradeType,
                        dealPrice,
                        leasePrice,
                        rentPrice,

                  FROM `aidepartners.aide.complex_real_price_prod`
                  WHERE deleteYn is null

                  ) AS price
                    ON area.complexNo = price.complexNo 
                      AND area.pyeongNo = price.areaNo

      WHERE danji.realEstateTypeCode = 'OPST' and (danji.sidoCode = '11' OR danji.sigunguCode IN ('41131','41133'))
  ) AS T1
)AS T2
WHERE T2.rn =1
"""
df = BH.read_table(sql)
```
