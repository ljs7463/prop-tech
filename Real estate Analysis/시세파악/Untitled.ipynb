```python
sql = f"""
SELECT *
FROM `aidepartners.aide.complex_danji_information` AS danji
LEFT JOIN (SELECT * FROM `aidepartners.aide.complex_area_information_prod`) AS area ON danji.complexNo = area.complexNo
LEFT JOIN (SELECT * FROM `aidepartners.aide.complex_real_price_prod`) AS price ON area.complexNo = price.complexNo AND area.pyeongNo = price.areaNo
WHERE danji.sigunguCode = '41287' 
      AND danji.realEstateTypeCode ='APT' 
      AND price.tradeType = 'A1'
      AND price.deleteYn is null 
      AND price.contractDate >='2019-01-01'
      AND areaSixGroupNo = 'A2'


"""
df = BH.read_table(sql)
```

```python
data= df.groupby(['complexNo','complexName','areaSixGroupNo','tradeYear','tradeMonth'])['dealPrice'].mean().reset_index()
```

```python
data['complexName'].unique()
```

```python
data['yearMonth'] = data['tradeYear'].astype(str) +'-'+data['tradeMonth'].astype(str)
data['yearMonth'] =pd.to_datetime(data['yearMonth'])
data['yearMonth']
```

```python
fix, ax = plt.subplots(figsize = (20, 6))
for i in list(data['complexName'].unique()):
    test =data[data['complexName']==i]
    sns.lineplot(test['yearMonth'], test['dealPrice'],linewidth = 2,color = 'gray',alpha = 0.4)
sns.lineplot(data['yearMonth'], data[data['complexName']=='탄현9단지일신']['dealPrice'],hue = data['complexName'],linewidth = 5)
ax.get_legend().remove()
## 1. x축 설정
ax.xaxis.set_ticks(pd.date_range('2019-01-01',"2022-12-01",freq = 'MS')) 
plt.xticks(rotation=270)

## 2. y축 설정
current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in current_values])

## 3. grid 설정
plt.grid(axis="y", c="lightgray")
plt.grid(axis="x", c="lightgray")

```
