```python
def vworld_api_continueline():
   

    # 엔드포인트
    end_point = "http://api.vworld.kr/req/data"

    # 파라미터
    service = "data"
    request = "GetFeature"
    data = "LP_PA_CBND_BUBUN"
    key = vworld_key
    emdCd = "11650106"


    # 요청 파라미터
    page = 1
    size = 1000
    attrFilter = f"emdCd:=:{emdCd}" # 잠원동
    
    featureCollection = {"type":"FeatureCollection"}
    features = []
    while True:
        # 요청 url
        url = f"{end_point}?service={service}&request={request}&data={data}&key={key}&attrFilter={attrFilter}&page={page}&size={size}"
        # api 요청
        res = json.loads(requests.get(url).text)
        # GeoJson 추가
        features = features + res["response"]["result"]["featureCollection"]["features"]
        # 페이지 설정
        page_current = int(res["response"]["page"]["current"])
        page_total = int(res["response"]["page"]["total"])
        page += 1
        if page_current == page_total:
            break
    featureCollection["features"] = features
    geo = featureCollection
    return geo
```

```python
vworld_api_continueline()
```

```python
# 줌 설정하기
zoom_size = 18

# folium 지도 생성하기
m = folium.Map(
    location=[37.51309296520273, 127.01616855318568],
    zoom_start=zoom_size
)

m
```

```python
# 배경지도 타일 설정하기
layer = "Hybrid"
tileType = "png"
tiles = f"http://api.vworld.kr/req/wmts/1.0.0/{vworld_key}/{layer}/{{z}}/{{y}}/{{x}}.{tileType}"
attr = "Vworld"

folium.TileLayer(
    tiles=tiles,
    attr=attr,
    overlay=True,
    control=True
).add_to(m)
m
```
